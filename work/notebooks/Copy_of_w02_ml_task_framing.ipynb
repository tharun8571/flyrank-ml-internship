## 3. Success metric

*One metric you can defend. What number means 'good'?*

# Success metric

The success metric is **Mean Absolute Error (MAE)**.

MAE measures the average difference between the predicted trend percentage and the observed trend percentage.

A lower MAE indicates that the model predicts content performance more accurately, leading to better content refresh decisions.

```python
print("Success Metric: Mean Absolute Error (MAE)")
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

# Why ML beats a fixed rule here

A fixed rule such as "refresh all pages older than 180 days" ignores many important factors that influence content performance.

Machine learning can analyze multiple features at the same time, including search volume, impressions, clicks, CTR, average position, sessions, and trend percentage.

By learning from historical data, the model can identify complex patterns and prioritize the pages that are most likely to benefit from a refresh. This provides better decision support than a simple rule-based approach.
